# Practical Lesson
Here we see the estimator of 2PCF for three points distribution; e.g., galaxies, temperature, etc.

We compare the distributions over a random distribution and look at the features of it.


## Introduction
Exist [many estimators](https://arxiv.org/abs/astro-ph/9912088v2); but the _natural one_ one may think of is the Peebles-Hauser \begin{equation*}\label{eq:natural}\tag{1}
\xi_\text{N} (r) = \dfrac{DD(r)-RR(r)}{RR(r)}
\end{equation*}
which is the comparisson between an histogram of data-data (two point data), $DD(r)$, and an histagram random-random, $RR(r)$. 

**However**, the usage of $\xi_\text{N} (r)$ is discouraged because one encounters biased reults if the survey sample area is relevant. An estimator poved to aid the above is _Landy-Szalay_ estimator
\begin{equation*}\label{eq:landaySzalay}\tag{2}
\xi_\text{LS}(r)=1+\dfrac{N_{RR}}{N_{DD}}\dfrac{DD_i(r)}{RR_i(r)}-2\dfrac{N_{RR}}{N_{DR}}\dfrac{DR_i(r)}{RR_i(r)}
\end{equation*}
where, $DR(r)$ is a mixup between data and randoms, $N_x$ is the number of counts of the histogram $x$. Even though there are many estimators, \eqref{eq:landaySzalay} encompasses our scope here and we employ it on the following. 

## Using example distributions
Let's define an histogram and the estimator

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# function to calculate distance between 2 points
def distance(ri= [],rj= []):
    tmpj=0.
    for ii in range(len(ri)):
        tmpj+= np.square(ri[ii]-rj[ii])
    return np.sqrt(tmpj)

def histogramDD(NBINS,RMIN,RMAX,DD=[],RR=[]):
    """Calculates an histogram from twho catalogs, these can have different dimensions.
    This receives number of bins, min range, max range and two arrays not needed to be of same lenght
    returns 1 array for bins, 1 array for histogram, the total sum of the histogram"""
    # define bin size
    binSize = (RMAX-RMIN)/NBINS;
    
    # define array of distances
    distances = [binSize*ii+RMIN for ii in range(NBINS+1)];
    
    # define array to store histogram
    histogram = np.zeros(NBINS+1)
    for ii in range(len(DD)):
        for jj in range(len(RR)):
            
            # calculate distance between points
            tmpDistance = distance(DD[ii],RR[jj])
            
            # check distance fits in range for histogram
            if RMIN > tmpDistance or tmpDistance > RMAX:continue
                
            # store count into histogram's bin
            histogram[int(round(tmpDistance/RMAX,3)*NBINS)] += 1
            
    return [distances,histogram,np.sum(histogram)]

# estimator Landy-Szalay
def estLS(BIN_MIN,BIN_MAX,DD= [],RR= [],DR= []):
    # define normalization constants 
    NDD = np.sum(DD[BIN_MIN:BIN_MAX])/np.sum(RR[BIN_MIN:BIN_MAX]);
    NDR = np.sum(DR[BIN_MIN:BIN_MAX])/np.sum(RR[BIN_MIN:BIN_MAX]);
    
    # define where to store estimator
    xi = [];
    
    # calculate and store estimator LS, lenghts of DD, RR and DR should be the same
    for ii in range(BIN_MIN,BIN_MAX):
        
        # some bins have zero value, so to fix xi's blow up in division we use try and except
        try:
            xi.append(1.+(1./NDD)*(DD[ii]/RR[ii])-(2./NDR)*(DR[ii]/RR[ii]))
            
        # if bin has zero value, we add just zero to xi
        except ZeroDivisionError:
            xi.append(0)
        
    return xi
```

Let's load the points distributions

```python
# load data from previously created distributions
disc=np.loadtxt("disc.dat")
rings=np.loadtxt("distri.dat")
randoms=np.loadtxt("ran.dat")
randoms2=np.loadtxt("ran2.dat")

# parameters for this data
boxSize=1000
NPTS=1000
```

```python
plt.figure(figsize=(12,3))
names = "disc rings randoms".split()
for ii,dist in enumerate([disc,rings,randoms]):
    plt.subplot(1,3,1+ii)
    plt.title(names[ii],fontsize=14)
    plt.xlim(0,boxSize);plt.ylim(0,boxSize)
    plt.plot(np.transpose(dist)[0],np.transpose(dist)[1],linestyle="none",marker=".",markersize=4);
plt.show()
```

define parameters for the estimator

```python
# number of bins
NBINS = 100

# minimal value for histogram
RMIN = 0

# maximal value for histogram
RMAX = boxSize*np.sqrt(2) # sqrt(2) because is a 2D box with equal lengh sides
```

Now let's calculate the histograms for data and randoms and calculate the correlation functions estimator

```python
# histograms for disc
DD_Disc = histogramDD(NBINS,RMIN,RMAX,disc,disc)
DR_Disc = histogramDD(NBINS,RMIN,RMAX,disc,randoms)
RR = histogramDD(NBINS,RMIN,RMAX,randoms,randoms)
#
# for rings
DD_Rings = histogramDD(100,0,boxSize*np.sqrt(2),rings,rings)
DR_Rings = histogramDD(100,0.,boxSize*np.sqrt(2),rings,randoms)
#
# for randoms
DD_Randoms = histogramDD(100,0.,boxSize*np.sqrt(2),randoms,randoms)
DR_Randoms = histogramDD(100,0.,boxSize*np.sqrt(2),randoms,randoms2)
RR_2 = histogramDD(100,0.,boxSize*np.sqrt(2),randoms2,randoms2)

# given we store histograms not centered in bins we draw 1-bin to left,
# so we calculate NBINS-1 values for estimator
rightEnd = NBINS-1

# calculate estimators
est_Disc=estLS(0,rightEnd,DD_Disc[1][:rightEnd],RR[1][:rightEnd],DR_Disc[1][:rightEnd])
est_Rings=estLS(0,rightEnd,DD_Rings[1][:rightEnd],RR[1][:rightEnd],DR_Rings[1][:rightEnd])
est_Randoms=estLS(0,rightEnd,DD_Randoms[1][:rightEnd],RR_2[1][:rightEnd],DR_Randoms[1][:rightEnd])
```

```python
plt.figure(figsize=(11,15))
names = "disc rings randoms".split()
estimators = [est_Disc,est_Rings,est_Randoms]
for ii,dist in enumerate([disc,rings,randoms]):
    plt.subplot(3,2,1+2*ii)
    plt.title(names[ii],fontsize=14)
    plt.xlim(0,boxSize);plt.ylim(0,boxSize)
    plt.plot(np.transpose(dist)[0],np.transpose(dist)[1],linestyle="none",marker=".",markersize=4);
    plt.subplot(3,2,(1+ii)*2)
    plt.title("$\\xi(r)$ "+names[ii],fontsize=14)
    plt.xlim(0,boxSize)
    plt.plot(DD_Disc[0][:-2],estimators[ii],linestyle="-",marker=".",markersize=4);
plt.show()
```
