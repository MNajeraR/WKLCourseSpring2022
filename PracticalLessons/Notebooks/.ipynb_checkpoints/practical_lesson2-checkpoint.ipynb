# Practical lesson (continuation) 
In this notebook we use a brute force algorithm for sake of understanding of the two point correlation function. Also we validate the obtained results with [TreeCorr code](https://rmjarvis.github.io/TreeCorr/_build/html/correlation2.html#) (TC from now on).

**Warning**: you need to have installed TC in order to successfully run the entire notebook. If you have not, you may look at [Mike Jarvis' github](https://github.com/rmjarvis/TreeCorr) or check on Mario's material for TC installation.

In the previous notebook we employ Landy-Szalay estimator to obtain the 2PCF $\xi_{\text{LS}}$, in this section we drop off the subindex for notation. Furthermore we notice that previously we compared 2 histograms, $DD$ and $RR$ explicitelly; this is clearer in Peebles-Hauser estimator, but when we have measurements this is rather strightforward. Let's say we have a catalog with convergence and other related quantities. This convergence value already overcame randomness and is unneeded to make a comparisson such as before. 

Then to calculate the 2PCF, let us start by import the libraries we employ along this notebook and define the needed functions

```python
import numpy as np # standard python module for numerical calculation
import matplotlib.pyplot as plt # standard python module for plotting
import treecorr # module to use TreeCorr's code

# define the correlation function - brute force

def corrKK(cat,rmin,rmax,nrbin):
        
    # function to calculate distance r
    def get_r(A=[],B=[]):
        return np.linalg.norm(np.array(A)-np.array(B))
    
    # function to get the bin for the estimator. It returns linear binning if inputed linear values,
    # logarithmic binning for inputed log values
    def binning(x,xmin,xBinsize):
        return int((x-xmin)/(xBinsize+10**(-5)))

    # initialize vars
    
    # define bin size in logarithm space
    rBinsize = np.log(rmax/rmin)/nrbin 
    
    # define array to store the correlation function 
    xikk = np.zeros(nrbin)
    
    # define array to count number of points in each bin; corresponds to the denominator in the above equation
    binCntr = xikk.copy()
    
    # calculate and store in xikk
    for ii,p1 in enumerate(cat):# catalog 1
        for jj,p2 in enumerate(cat):# catalog 2
            
            # avoid count twice over pairs
            if jj >= ii: continue
                
            # get the distance between pairs
            r=get_r(p1[:2],p2[:2])
            
            # check distance belongs to estimator range
            if r < rmin or r >= rmax:continue
            
            # if distance is within the relevant range store the correlation of points into bins
            xikk[binning(np.log(r),np.log(rmin),rBinsize)] += p1[-1]*p2[-1]
            binCntr[binning(np.log(r),np.log(rmin),rBinsize)] += 1  
            
    # normalize xikk
    for ii in range(nrbin):
        if binCntr[ii] == 0 :continue
        xikk[ii] = xikk[ii]/binCntr[ii]
            
    return [xikk,binCntr]
```

## Calculate the 2PCF
Now we have the basic way to calculate the 2PCF, we need a catalog whith positions and convergence. Below we define a toy catalog with $N=400$ and distances between $r=r_\text{min}=10^{-8}\text{ rad}$ and $r_\text{max}=boxsize = 0.2\text{ rad}$

```python
#number of points
N=400

# size of the box
boxsize=0.2

# distances min and max
rmin=10**(-8)
rmax=boxsize

# positions in right asenction and declination
np.random.seed(17)
ra=np.random.uniform(low=0,high=boxsize,size=N)
dec=np.random.uniform(low=0,high=boxsize,size=N)

# convergences k:[-1,1]
k=np.random.uniform(low=-1,high=1,size=N)

# define the catalog
minicat=np.transpose(np.reshape([ra,dec,k],(3,N)))
```

How does this catalog looks like? below we plot it; the intensity of the convergence goes from -1 to 1

```python
plt.figure(figsize=(5,4));
cm = plt.cm.get_cmap('cividis')#'copper')
sc=plt.scatter(ra,dec,c=k,s=20,cmap=cm,vmin=-1,vmax=1);plt.colorbar(sc);plt.xlim(0,boxsize);plt.ylim(0,boxsize);
plt.xlabel("ra [rad]",fontsize=14);plt.ylabel("dec [rad]",fontsize=14);plt.show()
```

Do you notice there is approximatelly the same dark points as bright ones? So the distribution seems to be uniform. Now let us run the code for the estimator, but first specify the number of desired bins in the distance between points

```python
# number of bins in r
nrbin=10
```

```python
%%time
xi,binCntr=corrKK(minicat,rmin,rmax,nrbin)
```

As you see, this seems midly fast. However, given the code structure it has a time function  $t(N)\sim N^2$. So it is very unconvenient for large $N$. We will check on this later but for now let us validate our result and watch it 

## Calculate the 2PCF Using TreeCorr
We use TC to validate our results; just bear in mind the results may differ because of TC's internal procedures. First we create a `Catalog` and `KKCorrelation` instances/objects of TC; we then use them to calculate the correlation function, _xi_

```python
# create the catalog using ra, dec and k from above. Units for ra and dec has to be specified
cat=treecorr.Catalog(ra=ra,dec=dec,k=k,ra_units='rad', dec_units='rad')
```

```python
# tell treecorr to create an instance of KKCorrelation Class, we call it tcKKcor 
tcKKcorr=treecorr.KKCorrelation(sep_units="rad",min_sep=rmin,max_sep=rmax,nbins=nrbin,verbose=2)
```

```python
# calculate the auto correlation xi for the catalog
tcKKcorr.process(cat)
```

We notice our method and TC have the same results; for the correlation function `xi` and `KKCorrelation.xi` and the number of assignation for each bin `binCntr` and `KKCorrelation.npairs`

```python
print("binCntr = ",np.array(binCntr,dtype="int"),"KKCorrelation.nparis",np.array(tcKKcorr.npairs,dtype="int"),sep="\n")
```

```python
plt.plot(tcKKcorr.rnom,tcKKcorr.xi,c="black",label="$\\xi_{TC}$")
plt.plot(tcKKcorr.rnom,xi,linestyle="none",marker=".",markersize=12,label="$\\xi_{BF}$",c="red")
plt.legend(frameon=False,fontsize=15);plt.xlabel("$r$")
plt.show()
```
